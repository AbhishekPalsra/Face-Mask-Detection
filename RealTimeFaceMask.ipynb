```python
def predict(img):
    input_image_resized = cv2.resize(img,(224,224))
    input_image_scaled = input_image_resized/255.0
    input_image_reshaped = np.reshape(input_image_scaled, (1,3,224,224))
    input_tensor = torch.tensor(input_image_reshaped,dtype=torch.float32).to(device)
    score = loaded_model(input_tensor).to(device)
    _,input_pred_label = score.max(1)
    return input_pred_label

```

```python
def draw_label(img,text,pos,bg_color):
    text_size=cv2.getTextSize(text,cv2.FONT_HERSHEY_SIMPLEX,1,cv2.FILLED)
    end_x=pos[0]+text_size[0][0]+2
    end_y=pos[1]+text_size[0][1]-2

    cv2.rectangle(img,pos,(end_x,end_y),bg_color,cv2.FILLED)
    cv2.putText(img,text,pos,cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1,cv2.LINE_AA)
```

```python
cap=cv2.VideoCapture(0)
```

```python
while True:
    ret,frame=cap.read()
    #call detection method
    y_pred=predict(frame)
    if y_pred==0:
        draw_label(frame,"No Mask",(30,30),(0,255,0))
    else:
        draw_label(frame,"Mask",(30,30),(0,0,255))
    cv2.imshow('window',frame)
    if cv2.waitKey(1) & 0xFF==ord('x'):
        break
cv2.destroyAllWindows()
```
